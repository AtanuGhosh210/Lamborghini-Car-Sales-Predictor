# Predicting Lamborghini Car Sales using Regression Analysis
This project uses supervised learning techniques to predict future sales growth of Lamborghini cars based on historical sales data.  
The goal is to build a linear regression model to accurately estimate the growth percentage given past sales data.  
**Dataset Source:** Kaggle (Lamborghini Sales and Stocks)  


```python
#Importing necessary libraries for data manipulation,visualization and model building
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
# Loading the Lamborghini Sales Dataset
data = pd.read_csv("Lambo-Future.csv")

# Displaying the first few rows of the dataset to understand the structure
print(data.head())
```

```python
# Rounding decimal values to 2 decimal places
data = data.round(2)

# Display the cleaned data
print(data.head(10))

# Checking for any missing or null values
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Scatter plot to visualize the relationship between Sales and Growth
plt.figure(figsize=(8, 5))
plt.scatter(data['Sales'], data['Growth'], color='blue')
plt.title('Lamborghini Sales vs Growth')
plt.xlabel('Sales')
plt.ylabel('Growth')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
X = data[['Sales']].values  # Independent variable (Sales)
y = data['Growth'].values   # Dependent variable (Growth)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

# Creating the linear regression model
model = LinearRegression()

# Training the model
model.fit(X_train, y_train)

# Predicting on the test data
y_pred = model.predict(X_test)
```

```python
plt.scatter(X, y, color='blue', label='Actual Growth')
plt.plot(X_test, y_pred, color='red', label='Predicted Growth', linewidth=2)
plt.title('Lamborghini Sales vs Growth')
plt.xlabel('Sales')
plt.ylabel('Growth')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score

# Evaluating the model
r2 = r2_score(y_test, y_pred)
print(f'R-squared score: {r2:.2f}')
```

```python
# Calculating Mean Absolute Error (MAE) And Mean Squared Error (MSE) For Evaluating Model's Performance
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

## Conclusion
The regression model successfully predicted Lamborghini sales with a high degree of accuracy.  
The R-squared score of **1.00** indicates an excellent fit, while the low MAE and MSE values suggest minimal prediction errors.  
This model can be useful for estimating future sales based on growth trends.
